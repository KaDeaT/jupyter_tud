# Взаимосвязь свойств ограниченности и устойчивости решений линейных систем

Решения неоднородной системы $$\dot{y} = A(t)y + f(t)$$ 
могут быть неограниченными функциями, но при этом система будет асимптотически устойчива.


Рассмотрим систему 
$$ \dot{y}  = - y + t + 1.$$

Тогда соответствующая ей однородная система (система в отклонениях):

$$ \dot{x} = -x.$$

Фундаментальная матрица системы в отклонениях $\Phi(t,t_0) = e^{t-t_0}$. 
  
  * $\Phi(t,t_0)$ ограничено, следовательно, по теореме 2 система в отклонениях устойчива;
  * $\Phi(t,t_0) \to 0$ при $t \to +\infty$. Тогда по теореме 3 система в отклонениях асимптотически устойчива.
  
При этом, решения исходной системы неограничены. Покажем это.

Решение неоднородной системы в форме Коши имеет вид:

$$ y(t,t_0,y_0) = \Phi(t,t_0) \left[ y_0 + \int_{t_0}^{t}{\Phi^{-1}(\tau,t_0) f(\tau) d\tau} \right]$$


В нашем случае получается:

$$ y(t,t_0,y_0) = e^{-(t-t_0)}\left[y_0 + \int_{t_0}^{t}{e^{(\tau-t_0)}(\tau + 1) d\tau } \right] = $$
$$ e^{-(t-t_0)}\left[y_0 + \int_{t_0}^{t}{\tau e^{(\tau-t_0)} d\tau } + \int_{t_0}^{t}{ e^{(\tau-t_0)} d\tau } \right] $$

Вычислим первый интеграл, входящий в правую часть.

$\int_{t_0}^{t}{\tau e^{(\tau-t_0)} d\tau } = 
\left|
\begin{array}{cc}
U = \tau & dV = e^{\tau - t_0}d\tau \\
dU = d\tau & V = e^{\tau - t_0}
\end{array}
\right| = \tau e^{(t-t_0)}\Bigr|_{t_0}^{t} - \int_{t_0}^{t}{ e^{(\tau-t_0)} d\tau } = te^{(t-t_0)} - t_0 - \int_{t_0}^{t}{ e^{(\tau-t_0)} d\tau  } $  

Продолжаем вычислять. 

$$ e^{-(t-t_0)}\left[y_0 + te^{(t-t_0)} - t_0 - \int_{t_0}^{t}{ e^{(\tau-t_0)} d\tau  } + \int_{t_0}^{t}{ e^{(\tau-t_0)} d\tau } \right] =   e^{-(t-t_0)}\left[y_0 + te^{(t-t_0)} - t_0 \right]   = t +  e^{-(t-t_0)}[y_0 - t_0] $$

Окончательно получаем:
$$ y(t,t_0,y_0) = t +  e^{-(t-t_0)}[y_0 - t_0] $$

Не трудно заметить, что все решения уходят на бесконечность при $t \to +\infty$.


Построим графики нескольких решений.

Красным цветом выделено невозмущённое решение неоднородной системы для задачи Коши $x(0) = 0$. Выпишем явно это решение:

$$ y(t,0,0) = t  $$

```python
import matplotlib.pylab as plt
import numpy as np

fig = plt.figure(figsize=(8, 8))
t = np.linspace(0, 4)
t0 = 0
x0 = 0
for x0 in range(-5, 5, 1):
    plt.plot(t, t + (x0-t0)*np.exp(-(t-t0)), color='blue' if x0 != 0  else 'red')
plt.xlabel(u"t, время",fontsize=13)
plt.show()
```

Может показаться, что выбранное частное решение является асимптотой для всех остальных решений. Но это верно лишь отчасти, т.к. любое частное решение будет асимптотой для всех остальных решений. (Это геометрическая интерпретация асимптотической устойчивости всей линейной системы). 

Построим ещё графики решений однородной системы.

```python
fig = plt.figure(figsize=(8, 8))
for x0 in range(-5, 5, 1):
    plt.plot(t, x0*np.exp(-(t-t0)), color='blue' if x0 != 0  else 'red')

plt.xlabel(u"t, время",fontsize=13)
plt.show()
```
