# Почему важен хороший базис

Рассмотрим систему $$\dot{x} = Ax,$$
где
$$ A = \left(
\begin{array}{cc}
2 & 3\\
1 & 4
\end{array}
\right)$$

Решения этой системы в исходном базисе выглядят следующим образом:

$$\left(
\begin{array}{c}
x_1\\
x_2 
\end{array}
\right)= \left(
\begin{array}{cc}
\frac{1}{4}e^{5t} + \frac{3}{4}e^{t} &  \frac{3}{4}e^{5t} - \frac{3}{4}e^{t}\\
\frac{1}{4}e^{5t} - \frac{1}{4}e^{t} & \frac{3}{4}e^{5t} + \frac{1}{4}e^{t}
\end{array}
\right) \left(
\begin{array}{c}
C_1\\
C_2 
\end{array}
\right)$$

Построим её фазовый портрет.

```python
import matplotlib.pylab as plt
import numpy as np
import odebook as ob

plt.figure(figsize=(8, 8))
ob.axes4x4(labels=('x','y'))

theta = np.linspace(0, 2*np.pi, 21 )
r = 2
x = np.cos(theta) * r
y = np.sin(theta) * r

ob.phaseportrait(lambda X, t=0: np.array([2*X[0]+3*X[1], 1*X[0]+4*X[1]]),
    np.array([x, y]).T, t=(-4, 2), n=50, linewidth=1.5)
 
#ob.vectorfield(np.arange(-4,4,1),np.arange(-4,4,1),lambda x,y: (2*x+3*y,x+4*y))
```

Выберем в качестве матрицы перехода 
$$S = \left(
\begin{array}{cc}
-3 & 1\\
1 & 1
\end{array}
\right).$$

Тогда $$S^{-1} = 
\frac{1}{4}\left(
\begin{array}{cc}
-1 & 1\\
1 & 3
\end{array}
\right)
$$

Матрица системы в новом базисе примет вид:

$$J_A = \left(
\begin{array}{cc}
1 & 0\\
0 & 5
\end{array}
\right) $$

Построим фазовое пространство.

```python
import matplotlib.pylab as plt
import numpy as np
import odebook as ob

plt.figure(figsize=(8, 8))
ob.axes4x4(labels=('x','y'))

theta = np.linspace(0, 2*np.pi, 21 )
r = 2
x = np.cos(theta) * r
y = np.sin(theta) * r

ob.phaseportrait(lambda X, t=0: np.array([X[0], 5*X[1]]),
    np.array([x, y]).T, t=(-4, 2), n=50, linewidth=1.5)
```

Общее решение этой системы в новом базисе выглядит следующим образом:

$$\left(
\begin{array}{c}
x_1\\
x_2 
\end{array}
\right)= \left(
\begin{array}{cc}
e^{t} & 0\\
0 & e^{5t}
\end{array}
\right) \left(
\begin{array}{c}
C_1\\
C_2 
\end{array}
\right)$$
